```julia
using Revise
push!(LOAD_PATH,"/scratch/Codes/RandomMeas_dev.jl/src/")
using ProgressBars
using Random
using ITensors
using PastaQ
using TimerOutputs
using RandomMeas
using StatsBase
```

```julia
N = 12
depth = 10
p = 0.01
noisemodel = (1 => ("depolarizing", (p = p,)),2 => ("depolarizing", (p = p,)))
ψ = Vector{MPS}()
ρ = Vector{MPO}()
ξ = siteinds("Qubit", N)
gates = buildcircuit(ξ,randomcircuit(N, depth=depth));
ψ0 = MPS(ξ,["Dn" for i in 1:N]);
ρ0 = outer(ψ0',ψ0)
for d in ProgressBar(1:depth, printing_delay=2)
        if d==1
            push!(ψ,apply(gates[d],ψ0;cutoff=1e-4))
            push!(ρ,apply(gates[d],ρ0;cutoff=1e-4,apply_dag=true))
        else
            push!(ψ,apply(gates[d],ψ[d-1];cutoff=1e-4))
            push!(ρ,apply(gates[d],ρ[d-1];cutoff=1e-4,apply_dag=true))
        end
        Apply_depo_channel!(ρ[d],ξ,p*ones(N))
end

println("state prepared in the `experiment")
println(maxlinkdim(ρ[depth]))
```

```julia
#Perform Randomized measurements
NM= 1000
XEB = zeros(Float64,depth)
selfXEB = zeros(Float64,depth)
XEB_c = zeros(Float64,depth)

data = zeros(Int8,(NM,N))
F = [real(inner(ψ[d]',ρ[d],ψ[d])) for d in 1:depth];

u = get_rotations(ξ,6)

for d in ProgressBar(1:depth, printing_delay=2)
    P0 = get_Born_MPS(ψ[d])
    truncate!(P0;cutoff=1e-4)
    
    orthogonalize!(ρ[d],1)
    get_RandomMeas_MPO!(data,ρ[d],u,NM)

    for m in 1:NM
        V = ITensor(1.)
        for j=1:N
              V *= (P0[j]*state(ξ[j],data[m,j]))
        end
        XEB[d] += 2^N/NM*real(V[])-1/NM
    end

    selfXEB[d] = 2^N*real(inner(P0,P0))-1
    XEB_c[d] = XEB[d]/selfXEB[d]
end
```

```julia
using PyPlot
plot(1:depth,F,"-o")
plot(1:depth,XEB,"-o")
plot(1:depth,XEB_c,"-o")
xlabel("depth")
ylabel("Fidelity")
yscale("log")
legend(["fidelity","XEB","XEB/selfXEB"],loc=0)
```
